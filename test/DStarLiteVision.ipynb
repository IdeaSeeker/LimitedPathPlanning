```python
den204d_filename = '../data/den204d.map'
den312d_filename = '../data/den312d.map'
Paris_256_filename = '../data/Paris_1_256.map'
```

```python
def run_on(map_filename, start, finish, pixel_size = 20, gif_speed = 200, vision_distance = 1):
    _map = Map()
    _map.read_from_string(*read_map(map_filename))

    d_star = DStarLite(_map, start, finish, vision_distance=vision_distance)
    d_star.run()

    gif_filename = f'd_star_lite_{start.i}_{start.j}_{finish.i}_{finish.j}_vision_{vision_distance}.gif'
    draw_path(_map, d_star._path, gif_filename, pixel_size, gif_speed, vision_distance)

    return Img(filename=gif_filename)
```
