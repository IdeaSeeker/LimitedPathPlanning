```python
def run_on(map_filename, start, finish):
    _map = Map()
    _map.read_from_string(*read_map(map_filename))

    d_star = LimitedDStar(_map, start, finish)
    d_star.run(map_logging=True)

    gif_filename = f'd_star_{start.i}_{start.j}_{finish.i}_{finish.j}.gif'
    draw_path(_map, d_star._path, gif_filename, d_star._log_current_map)

    return Img(filename=gif_filename)
```
